## ü§ñ What is ROS?

<img src="https://www.ros.org/imgs/logo-white.png" width="480" alt="ROS Logo" href="https://www.ros.org/">

**ROS** is a collection of software libraries and tools for building robot applications. It lets different programs (called nodes) communicate to control robots. ROS is **open source**, **free to use**, and **platform-independent**. ROS 2 is the latest version with improved performance and security.

> **‚ö° ROS Reinvents the wheel so you don't have to!**
>
> Learn more: [ROS Introduction](http://wiki.ros.org/ROS/Introduction)

**Key Features:**
- üîå **Distributed:** Nodes talk over the network
- üõ†Ô∏è **Tools:** Command-line and graphical tools for debugging
- üêç **Multi-language:** Python, C++, and more
- üì¶ **Community:** Huge ecosystem of packages
- üîì **Open Source:** Free BSD license

## üîó ROS Concepts

| Concept | What it is |
| --- | --- |
| **Nodes** | Programs that communicate with each other |
| **Topics** | Named channels where nodes send/receive messages |
| **Messages** | Data payloads sent over topics |
| **Services** | Request-response communication (like a function call) |
| **Actions** | Long-running tasks with feedback |
| **Parameters** | Global configuration values |

## üíª ROS Command-line Tools

Common tools for managing and debugging ROS systems:

### ros2 run

Run an executable from a package:

```bash
# Run turtlesim node
ros2 run turtlesim turtlesim_node

# Press ctrl+c to stop
```

#### ros2 launch

The `ros2 launch` command allows you to run multiple ROS nodes with a single command. It also allows you to set parameters on the nodes, and to specify remapping rules for topics. For example, to launch the turtlesim_node from the turtlesim package, you can use the following command:

```bash
# stop the turtlesim_node if it is running
# in a new terminal run turtlesim multi sim using roslaunch
ros2 launch turtlesim multisim.launch.py
```

### ros2 node

The `ros2 node` command allows you to get information about running ROS nodes. For example, to get a list of all running nodes, you can use the following command:

```bash
# in a new terminal run the following command
ros2 node list
```

#### ros2 topic

The `ros2 topic` command allows you to get information about ROS topics. For example, to get a list of all topics, you can use the following command:

```bash
# in the same terminal run the following command
ros2 topic list
```

#### ros2 interface (msg)

The `ros2 interface` command allows you to get information about ROS messages. For example, to get a list of all messages, you can use the following command:

```bash
# in the same terminal run the following command
ros2 interface list
```

#### Combine nodes to create a robot application

Run both the turtlesim and the turtle_teleop_key nodes to control the turtle in the turtlesim simulator.

```bash
# in a new terminal run turtlesim_node
ros2 run turtlesim turtlesim_node

# in another new terminal run turtle_teleop_key
ros2 run turtle_teleop_key turtle_teleop_key
```

![turtlesim](./docs/images/turtlesim.png)

#### Use `ros2 topic pub` to publish messages to move the turtle

```bash
# in a new terminal run the following command to move the turtle up
ros2 topic pub /turtle1/cmd_vel geometry_msgs/msg/Twist "{linear: {x: 1.0, y: 0.0, z: 0.0}, angular: {x: 0.0, y: 0.0, z: 1.0}}"
```

The turtlesim should drive around in a circle.

## ROS Graphical Tools

- [**RQT**](<https://wiki.ros.org/rqt>) is a series of small graphical tools using QT for visualizing and interacting with a ROS system. RQT plugins are available to visualize topics, services, actions, parameters, and more.
- [**RViz**](<https://wiki.ros.org/rviz/UserGuide>) is a 3D visualization tool for ROS. RViz allows you to visualize sensor data, robot models, and other information.
- [**Gazebo**](<https://gazebosim.org/home>) is a 3D simulation environment for ROS. Gazebo allows you to simulate robots in complex indoor and outdoor environments.
