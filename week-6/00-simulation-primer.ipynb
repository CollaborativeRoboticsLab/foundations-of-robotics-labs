## ðŸŽ® Gazebo Simulator

In this course we will use the [Gazebo simulator](http://gazebosim.org/). Gazebo is a 3D simulator that allows us to simulate robots and environments. It is very powerful and can simulate many different types of robots and sensors. It is also open source and has a large community of users and developers.

### Gazebo and ROS

Gazebo is often used together with ROS and it is the default simulator in ROS. ROS provides a set of tools to interact with Gazebo.

**Try it out!**

```bash
# export turtlebot3 model
export TURTLEBOT3_MODEL=burger

# Start a Gazebo-ROS simulation with TurtleBot 3
ros2 launch turtlebot3_gazebo turtlebot3_world.launch.py
```

![Gazebo Sim](./docs/images/ign-gazebo.png)

### The Robot View (RViz)

The simulation shows the robot and its environment, but to see what the robot "sees" we use RViz. RViz is a visualization tool that allows us to see the robot's sensors and state.

> **Try it out!**
>
> make sure you have the Gazebo simulation running before starting RViz, and run the following command in a new terminal:

```bash
# Start RViz
rviz2
```

Try to add the following displays in RViz:
- TF
- RobotModel
- LaserScan

It should look something like this:

#### ROS Network

Use the ROS Command Line Tools to check the ROS network. The following commands can be useful:

```bash
# List all active ROS nodes
ros2 node list

# List all active ROS topics
ros2 topic list

# Get information about a specific topic
ros2 topic info /topic_name

# Get information about a specific node
ros2 node info /node_name
```
