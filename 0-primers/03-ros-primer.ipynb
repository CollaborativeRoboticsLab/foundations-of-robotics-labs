## What is ROS?

`ROS` is a set of *software libraries and tools* that help you build robot applications. ROS is a *distributed framework* of processes (`nodes`) that work together passing information to complete robot tasks. These processes can be grouped into packages and distributed through a network of computers.

***ROS Reinvents the wheel so you don't have to!***

ROS is a collection of tools, libraries, and conventions to `simplify` creating complex and robust robot behavior for any robot. ROS provides a *client library* that helps write programs in a variety of `programming` languages. ROS also provides a set of tools that enable you to easily create, manage, and visualize the data flowing between programs. ROS is completely **open source** (BSD) and free to use, change, and commercialize upon.

ROS is designed to be as platform-independent as possible. ROS is not an operating system in the traditional sense of process management and scheduling; rather, you can think of ROS as a meta-operating system that provides services on top of an existing operating system. ROS is similar to *'robot frameworks'* such as Player, YARP, Orocos, CARMEN, Orca, MOOS, and Microsoft Robotics Studio.

> The ROS wiki contains a more detailed [introduction](http://wiki.ros.org/ROS/Introduction) to ROS.

`ROS2` is the latest version of ROS, which includes improvements in performance, security, and support for real-time systems. In these labs, we will be using ROS2.

Key features include:

* **Distributed Communication:** Nodes communicate via various methods.
* **Tools:** Command-line and graphical tools simplify running, visualizing, and interacting with nodes.
* **Language Independence:** Supports multiple languages (C++, Python, etc.).
* **Debugging Support:** Tools and libraries aid in visualizing, debugging, and testing.
* **Open Source:** Free for commercial and research use (BSD license).
* **Community & Ecosystem:** Large and active community with extensive packages and resources.
* **Extensible:** Tools for creating and managing new packages.
* **Robotics Data Types:** Standardized data types for common robotics concepts.
* **Data Distribution:**  Uses topics and publish/subscribe for efficient data exchange.

## ROS Concepts

| Concept | Description |
| --- | --- |
| **Nodes** | A `node` is an executable that uses ROS to communicate with other nodes. |
| **Topics** | A `topic` is a named bus over which nodes exchange messages. Topics have anonymous publish/subscribe semantics, which decouples the production of information from its consumption. Nodes that are interested in data subscribe to the relevant topic; nodes that generate data publish to the relevant topic. There can be multiple publishers and subscribers to a topic. |
| **Messages** | `Messages` are the payload of information that is sent and received via topics. Standard primitive types (integer, floating point, boolean, etc.) are supported, as are arrays of primitives. Messages can include nested types. |
| **Services** | A `service` is like a topic but it is transactional. When a service is requested there is a response. Then the services is complete. |
| **Actions** | An `action` is a special type of communication pattern for robotics combining a service and a topic. Actions are useful for goal(service) oriented tasks that may take a long time to complete, and for which feedback(topic) is useful. |
| **Parameters** | A `parameter` is a piece of information that might be important to many nodes. It can be set globally and read by any node, which helps keep track of the state of the system. |

## ROS Command-line Tools

ROS provides tools for working with robots. Command line tools are useful for managing the ROS environment, debugging, and testing. Common command line tools include:

#### ros2run

The `ros2 run` command allows you to run an executable in an arbitrary package without having to give its full path. For example, to run the turtlesim_node from the turtlesim package, you can use the following command:

```bash
# in a new terminal run turtlesim_node
ros2 run turtlesim turtlesim_node

# ctrl+c to stop the node
# describe what you observed when you ran the command
```

#### ros2launch

The `ros2 launch` command allows you to run multiple ROS nodes with a single command. It also allows you to set parameters on the nodes, and to specify remapping rules for topics. For example, to launch the turtlesim_node from the turtlesim package, you can use the following command:

```bash
# stop the turtlesim_node if it is running
# in a new terminal run turtlesim multi sim using roslaunch
ros2 launch turtlesim multisim.launch.py
```

### ros2node

The `ros2 node` command allows you to get information about running ROS nodes. For example, to get a list of all running nodes, you can use the following command:

```bash
# in a new terminal run the following command
ros2 node list
```


#### ros2topic

The `ros2 topic` command allows you to get information about ROS topics. For example, to get a list of all topics, you can use the following command:

```bash
# in the same terminal run the following command
ros2 topic list
```

#### ros2msg
The `ros2 msg` command allows you to get information about ROS messages. For example, to get a list of all messages, you can use the following command:

```bash
# in the same terminal run the following command
ros2 msg list
```

#### Combine nodes to create a robot application

Run both the turtlesim and the turtle_teleop_key nodes to control the turtle in the turtlesim simulator.

```bash
# in a new terminal run turtlesim_node
ros2 run turtlesim turtlesim_node

# in another new terminal run turtle_teleop_key
ros2 run turtle_teleop_key turtle_teleop_key
```

## ROS Graphical Tools

- [**RQT**](<https://wiki.ros.org/rqt>) is a series of small graphical tools using QT for visualizing and interacting with a ROS system. RQT plugins are available to visualize topics, services, actions, parameters, and more.
- [**RViz**](<https://wiki.ros.org/rviz/UserGuide>) is a 3D visualization tool for ROS. RViz allows you to visualize sensor data, robot models, and other information.
- [**Gazebo**](<https://gazebosim.org/home>) is a 3D simulation environment for ROS. Gazebo allows you to simulate robots in complex indoor and outdoor environments.
